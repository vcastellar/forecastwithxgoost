```python
# Create lagged features for 'PJME_MW' up to 7 lags
for i in range(1, 8):
    df[f'PJME_MW_lag{i}'] = df['PJME_MW'].shift(i)

# Display the first few rows to show the new lagged features
display(df.head(10))
```

```python
df.loc[df.index >'2015-12-31']
```

```python
# Define the split date
split_date = '2015-12-31'

# Split the DataFrame into training and testing sets
df_train = df.loc[df.index <= split_date]
df_test = df.loc[df.index > split_date]

print(f"Shape of training set: {df_train.shape}")
print(f"Shape of testing set: {df_test.shape}")

print("\nFirst 5 rows of training set:")
display(df_train.head())

print("\nFirst 5 rows of testing set:")
display(df_test.head())
```

```python
import xgboost as xgb

# Define features (X) and target (y)
FEATURES = ['Month', 'DayofWeek', 'Hour', 'PJME_MW_lag1', 'PJME_MW_lag2', 'PJME_MW_lag3', 'PJME_MW_lag4', 'PJME_MW_lag5', 'PJME_MW_lag6', 'PJME_MW_lag7']
TARGET = 'PJME_MW'

# Combine features and target for training set to drop NaNs synchronously and sort the index
train_df_cleaned = df_train[FEATURES + [TARGET]].dropna().sort_index()
X_train = train_df_cleaned[FEATURES]
y_train = train_df_cleaned[TARGET]

# Combine features and target for test set to drop NaNs synchronously and sort the index
test_df_cleaned = df_test[FEATURES + [TARGET]].dropna().sort_index()
X_test = test_df_cleaned[FEATURES]
y_test = test_df_cleaned[TARGET]

print(f"Shape of X_train after dropping NaNs and sorting: {X_train.shape}")
print(f"Shape of y_train after dropping NaNs and sorting: {y_train.shape}")
print(f"Shape of X_test after dropping NaNs and sorting: {X_test.shape}")
print(f"Shape of y_test after dropping NaNs and sorting: {y_test.shape}")

# Initialize and train the XGBoost Regressor model
model = xgb.XGBRegressor(n_estimators=2000, early_stopping_rounds=50,
                         learning_rate=0.01, random_state=42)

model.fit(X_train, y_train,
          eval_set=[(X_train, y_train), (X_test, y_test)],
          verbose=False)

print("Modelo XGBoost entrenado con éxito.")
```

```python
# Display the best score and best iteration from the trained model
print(f"Best score: {model.best_score}")
print(f"Best iteration: {model.best_iteration}")

# You can also inspect other attributes of the model object, for example:
print(model.get_params())
print(model.feature_importances_)
```

```python
results = model.evals_result()
epochs = len(results['validation_0']['rmse'])
x_axis = range(0, epochs)

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x_axis, results['validation_0']['rmse'], label='Train RMSE')
ax.plot(x_axis, results['validation_1']['rmse'], label='Test RMSE')
ax.legend()
plt.ylabel('RMSE')
plt.title('XGBoost RMSE durante el entrenamiento')
plt.grid(True)
plt.show()

print(f"Mejor iteración: {model.best_iteration}")
print(f"Mejor RMSE en el conjunto de prueba: {model.best_score:.2f}")
```

### Búsqueda de Cuadrícula (Grid Search) para Optimización de Hiperparámetros

Vamos a definir un `param_grid` con algunas opciones para `n_estimators`, `learning_rate` y `max_depth`. Luego, usaremos `GridSearchCV` para encontrar la mejor combinación.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Rendimiento del modelo en la muestra de prueba (después de ordenar los índices):")
print(f"  Mean Absolute Error (MAE): {mae:.2f}")
print(f"  Mean Squared Error (MSE): {mse:.2f}")
print(f"  Root Mean Squared Error (RMSE): {rmse:.2f}")
```
